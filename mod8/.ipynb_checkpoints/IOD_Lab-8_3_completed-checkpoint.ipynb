# Step 7: Split the Data

Need to check for **Supervised** modelling:
- Number of known cases or observations
- Define the split in Training/Test or Training/Validation/Test and their proportions
- Check for unbalanced classes and how to keep or avoid it when spliting

```python
X = data.drop(columns = ['id', 'diagnosis'])
y = data['diagnosis']

y = y.to_numpy()
ohc = OneHotEncoder(sparse = False)
y_ohc = ohc.fit_transform(y.reshape(-1,1))

# 0 is malignant, 1 is benign
y = np.zeros(len(data))

for i in range(len(y)):
    y[i] = y_ohc[i][0]
    
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size  =0.33)
```

# Step 8: Define and Fit Models

Define the model and its hyper-parameters.

Consider the parameters and hyper-parameters of each model at each (re)run and after checking the efficiency of a model against the training and test datasets.

```python
seed = 32
nfolds = KFold(n_splits = 10, random_state = seed, shuffle = True)
svm = SVC()
svm.fit(X_train, y_train)
svm_cv = cross_val_score(svm, X_train, y_train, cv = nfolds)

logreg = LogisticRegression(max_iter = 5000)
logreg.fit(X_train, y_train)
lg_cv = cross_val_score(logreg, X_train, y_train, cv = nfolds)

nb = GaussianNB()
nb.fit(X_train, y_train)
nb_cv = cross_val_score(nb, X_train, y_train, cv = nfolds)

knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
knn_cv = cross_val_score(knn, X_train, y_train, cv = nfolds)
```

```python
classifiers = list()
classifiers.append(('knn', knn))
classifiers.append(('logreg', logreg))
classifiers.append(('nb', nb))

model = StackingClassifier(estimators = classifiers, final_estimator = logreg )
model.fit(X_train, y_train)
```

```python
seed = 9
nfolds = KFold(n_splits = 10, random_state = seed, shuffle = True)
model_result = cross_val_score(model, X_train, y_train, cv = nfolds)
```

# Step 9: Verify and Evaluate the Training Model
- Use the **training** data to make predictions
- Check for overfitting
- What metrics are appropriate for the modelling approach used
- For **Supervised** models:
    - Check the **Training Results** with the **Training Predictions** during development
- Analyse, modify the parameters and hyper-parameters and repeat (within reason) until the model does not improve

```python
y_train_pred = model.predict(X_train)
```

```python
train_tn, train_fp, train_fn, train_tp = metrics.confusion_matrix(y_train, y_train_pred).ravel()
train_cm = pd.DataFrame(index = ['Predicted Positive', 'Predicted Negative'], columns = ['Actual Positive', 'Actual Negative'])
train_cm.iloc[0,0] = train_tp
train_cm.iloc[0,1] = train_fp
train_cm.iloc[1,0] = train_fn
train_cm.iloc[1,1] = train_tn
train_cm
```

```python
accuracy_score = metrics.accuracy_score(y_train, y_train_pred)
accuracy_score
```

```python
recall_score = metrics.recall_score(y_train, y_train_pred)
recall_score
```

```python
fpr, tpr, _ = metrics.roc_curve(y_train, y_train_pred)
train_rocauc = metrics.auc(fpr,tpr)
```

```python
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange', lw = lw, label = 'ROC curve (area = %0.2f)' % train_rocauc)
plt.plot([0, 1], [0, 1], color = 'navy', lw = lw, linestyle = '--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc = "lower right")
plt.show()
```
