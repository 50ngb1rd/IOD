### Sector analysis 

#### Creating sector datasets -> df_sector

```python
# create list for the loop
sector_list = df['Sector'].unique()
date_list = df['Date'].unique()

# used to store the total volume for each sector each day
df['Total Volume per Sector'] = 0


# creating a dataset grouped by date and sector with the sum of volume for each sector
df_sector_vol = df.groupby(['Date','Sector']).Volume.sum()
df_sector_vol = df_sector_vol.reset_index()
```

```python
# loop to populate the total volume for each sector each day in the original dataframe 
for i in date_list:
    df_s = df_sector_vol[df_sector_vol['Date'] == i]
    for j in sector_list:
        vol = df_s[df_s['Sector'] == j].iloc[0]['Volume']
        selection = (df['Date'] == i)  & (df['Sector'] == j)
        df.loc[selection,['Total Volume per Sector']] = vol
```

```python
# calculate weighted returns for each different type of returnin the dataset
df['Volume Weights'] = df['Volume'] / df['Total Volume per Sector'] 
df['w_daily_return'] = df['daily_return'] * df['Volume Weights']
df['w_log_returns'] = df['log_return'] * df['Volume Weights']
df['w_ytd_returns'] = df['YTD_return'] * df['Volume Weights']
```

```python
# by summing the daily weighted return for each sector, we get the average weighted return for each sector in each day
df_sector = df.groupby(['Date','Sector']).agg({'w_daily_return':'sum','w_log_returns':'sum','w_ytd_returns':'sum'})
df_sector = df_sector.reset_index()
```

```python
# when performing min and max, 0 return was causing issues therefore deleted. Reason is only the first date of everys stock had a return of 0
df_sector_del = df_sector[df_sector['Date'] == '2020-01-22']
df_sector = df_sector.drop(df_sector_del.index)
```

```python
sp_500_2019 = pd.read_csv('E:/IOD_data/sp_500_2019.csv')
sp_500_11_19 = pd.read_csv('E:/IOD_data/sp_500_11_19.csv')
```

```python
sp_500_2019['log_ytd_return'] = np.log(sp_500_2019['ytd_return'] + 1)
sp_500['log_ytd_return'] = np.log(sp_500['ytd_return'] + 1)
```

```python
# for daily returns, running for whole time period
m_dr_2020 = np.mean(sp_500['Daily return'])
m_dr_2019 = np.mean(sp_500_2019["Daily return"])
std_dr_2019 = np.sqrt(np.var(sp_500_2019['Daily return']))
z_dr =  (m_dr_2020 - m_dr_2019) / std_dr_2019
z_dr_p_value = (1-stats.norm.cdf(z_dr))
print(z_dr_p_value)
print(m_dr_2020, m_dr_2019)
```

```python
# for YTD returns, proof for one
# for close prices, 2019 dataset only
m_ytd_2020 = np.mean(sp_500['log_ytd_return'])
m_ytd_2019 = np.mean(sp_500_2019['log_ytd_return'])
std_ytd_2019 = np.sqrt(np.var(sp_500_2019['log_ytd_return']))
z_ytd =  (m_ytd_2020 - m_ytd_2019) / std_ytd_2019
z_ytd_p_value = (1- stats.norm.cdf(abs(z_ytd)))
print(z_ytd_p_value)
print(m_ytd_2020, m_ytd_2019)
```

```python
def lr(x,y):
    y = y 
    x_mean = x.mean()
    y_mean = y.mean()

  
    x_diff = (x - x_mean)
    y_diff = (y - y_mean)
    xy_diff = x_diff * y_diff
    x_diff_sq = np.square(x_diff)
    x_diff_sum = x_diff_sq.sum()
    xy_diff_sum = xy_diff.sum()
 

    b1 = xy_diff_sum / x_diff_sum


    
    b0 = y_mean - (b1 * x_mean)

    

    y_hat = b0 + (b1 * x)

    
    res = y - y_hat
    
    ss_res = np.square(res).sum()
    ss_total = np.square(y - y_mean)
    ss_total = ss_total.sum()
    r_sq = 1 - (ss_res / ss_total)

    
    return r_sq, b1, b0

    
```
