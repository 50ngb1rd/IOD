#### 2.C [BONUS] Write A Function to Row Count



```
def row_count(column_name):
    number_of_rows = 0
    return number_of_rows
```



````python
def row_count(column_name):
    d_store = {}
    for i in range(len(df[column_name])):
        if df[column_name][i] in d_store == False:
            d_store[df[column_name][i]] = 1
        else:
            d_store[df[column_name][i]] += 1
            
````
