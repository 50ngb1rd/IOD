# Assingment 1 : Understanding the impact of COVID-19 cases (within US vs outside US) on investor sentiments in the S&P 500

In early 2020, the rise of COVID-19 cases has been reported to cause a drastic change in  investing patterns, leading to changes in stock prices and trading volumes. For example, during an increase in COVID cases, it was reported that stocks within the travel industry fell, and stocks within the tech industry rose. <br>


Drastic changes in  investing patterns are often related to changed investor sentiments for a particular industry/stock due external market events. Although COVID-19 is one such market event, it still unknown if local stock markets within market are more sensitive to global case numbers or more localised case numbers. With this in mind, this notebook will attempt to answer the below <b>business question <b>:
<br><br>
<b><i> How does the change in global vs local COVID cases impact investor sentiment relayed by investing patterns and what is the resultant impact on the stock market?

```python
stock_ticker = sp500['Symbol']
stock_ticker = stock_ticker.tolist()

master = pd.DataFrame(columns = ['Open', 'High','Low', 'Close', 'Volume', 'Dividends', 'Stock Splits', 'Symbol', 'Sector', 'Industry'])

for i in stock_ticker:
    stock = yf.Ticker(i)
    x = stock.history(start = '2020-01-22', end = '2021-12-06')
    y = stock.info
    x['Symbol'] = i
    x['Sector'] = y['sector']
    x['Industry'] = y['industry']
    
    master = pd.concat([x, master])
```

```python
master_list = master.reset_index()
master_list
```

```python
master_list = master_list.rename(columns = {'index':'Date','Company': 'Symbol'})
master_list
```

```python
master_stock_list = pd.merge(master_list,sp500, on ='Symbol')
master_stock_list = master_stock_list.drop(columns = 'Adj Close')
```
