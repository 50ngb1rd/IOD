## Using a webcam through OpenCV

import the necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
```

below code block uses your default camera to capture a video. Pressing Q will take the picture and quit OpenCV

```python
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
vid.release()
cv2.destroyAllWindows()
```

Below will display the pciture that you have taken

```python
fig, ax = plt.subplots()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
ax.imshow(frame)
```
